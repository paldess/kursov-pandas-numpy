удаление целевой переменной из данных для обучения

```python
data_train = data.drop('Price', axis=1)
data_train
```

замена пустых значений в жилой прлощади, на среднее значения площади комнаты умноженное на кол-во комнат

```python
data_train['LifeSquare'] = data_train['LifeSquare'].fillna(0)
price_one_room = data_train['LifeSquare'][data_train['LifeSquare']!=0].mean()/data_train['Rooms'].mean()
price_one_room
```

```python
data_train['LifeSquare'][data_train['LifeSquare']==0] = price_one_room * data_train['Rooms']
```

```python
data_train['Ecology_2'].unique()
```

```python
data_train['Ecology_3'].unique()
```

```python
data_train['Ecology_2'][data_train['Ecology_2']=='B'] = 0
data_train['Ecology_2'][data_train['Ecology_2']=='A'] = 1
data_train['Ecology_3'][data_train['Ecology_3']=='B'] = 0
data_train['Ecology_3'][data_train['Ecology_3']=='A'] = 1
data_train
```

```python
data_train = data_train.astype({'Ecology_2': int, 'Ecology_3': int})
```

### Обучение модели

```python
from sklearn.ensemble import RandomForestRegressor
x = 20
regres = RandomForestRegressor(max_depth=x, random_state=42)
```

```python
regres.fit(data_train, target)
```

подготовка тестовых данных 

```python
data_test = pd.read_csv('test.csv')
```

```python
# data_test.set_index('Id', inplace=True)
data_test['Shops_2'][data_test['Shops_2']=='A'] = 1
data_test['Shops_2'][data_test['Shops_2']=='B'] = 0
data_test['Healthcare_1'] = data_test['Healthcare_1'].fillna(0)
data_test['Ecology_2'][data_test['Ecology_2']=='B'] = 0
data_test['Ecology_2'][data_test['Ecology_2']=='A'] = 1
data_test['Ecology_3'][data_test['Ecology_3']=='B'] = 0
data_test['Ecology_3'][data_test['Ecology_3']=='A'] = 1
data_test['LifeSquare'] = data_test['LifeSquare'].fillna(0)
data_test['LifeSquare'][data_test['LifeSquare']==0] = price_one_room * data_test['Rooms']
```

```python
data_test
```
